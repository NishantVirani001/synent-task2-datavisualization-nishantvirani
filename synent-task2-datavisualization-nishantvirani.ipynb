#Task 2: Data Visualization

**Problem Statement :**

- The objective of this project is to visualize the Iris dataset and identify relationships between different flower species and their characteristics.

**Dataset Details :**

- The Iris dataset contains sepal length, sepal width, petal length, petal width, and flower species.

**Approach :**

- I created bar charts, histograms, and scatter plots to compare different features and understand the distribution of the data.

**Dataset Loading**

```python
import pandas as pd
import numpy as np

df = pd.read_csv("iris.csv")
print(df)
df.info()
print(df.head())
print(df.shape)
print(df.describe())
```

##Bar Chart

- This bar chart shows the number of flowers belonging to each species in the Iris dataset. It helps compare the frequency of each category.

```python
import matplotlib.pyplot as plt
count = df['species'].value_counts()
print(count)

count.plot(kind='bar')

plt.title("Species Count")
plt.xlabel("Species")
plt.ylabel("Count")

plt.show()
```

##Histogram

- The histogram displays the distribution of sepal lengths. It helps understand the spread and frequency of the measurements.

```python
df['sepal_length'].plot(kind='hist')

plt.title("Sepal Length Distribution")
plt.ylabel("count")
plt.xlabel("sepal length")

plt.show()
```

##Scatter Plot

- This scatter plot shows the relationship between sepal length and petal length. Each point represents one flower sample.

```python
plt.scatter(df['sepal_length'],df['petal_length'])

plt.title("Sepal length VS Petal length")
plt.ylabel("Petal length")
plt.xlabel("Sepal langth")

plt.show()
```

##Compare features


- Scatter plot shows relationship between:

- Sepal width

- Petal width

```python
plt.scatter(df['sepal_width'],df['petal_width'])

plt.title("Sepal Width VS Petal Width")
plt.ylabel("Petal Width")
plt.xlabel("Sepal Width")

plt.show()
```
